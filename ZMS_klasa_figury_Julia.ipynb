```julia
abstract type Figura end

mutable struct Prostokąt
    dlugosci::Vector{Int64}
    Prostokąt() = new([1,2,2,1])
end

function obwod( prost::Figura )
    return sum(prost.base.dlugosci)       
end

function pole( prost::Figura )
    return minimum(prost.base.dlugosci)*maksimum(prost.base.dlugosci)   
end

mutable struct Kwadrat <: Figura
    base::Prostokąt 
    dlugosci::Vector{Int64}
    Kwadrat( dlugosci::Vector{Int64} ) = new( Prostokąt(),  dlugosci)
end

function obw( kw::Kwadrat )
    return sum(kw)
end

function pole( kw::Kwadrat )
    return kw[1]^2
end
```
