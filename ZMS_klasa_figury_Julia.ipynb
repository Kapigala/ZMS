```julia
abstract type Figura end

mutable struct Prostokąt <: Figura
    dlugosci::Vector{Int64}
    Prostokąt(dlugosci=[2,3,3,2]) = new(dlugosci)
end

function obwod( fig::Prostokąt )
    return sum(fig.dlugosci)
    end
function pole( fig::Prostokąt )
    return maximum(fig)*minimum(fig)  
end

mutable struct Kwadrat <: Figura
    kw(dlugosci=[1,1,1,1]) = new(dlugosci)
end
function obwod( kw::Kwadrat )
    return 4*kw[1]
end
function pole( fig::Kwadrat )
    return kw[1]^2 
end
```
